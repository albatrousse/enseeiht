## Justifications

1. Soit la fonction $f_0 : \mathbb{R} \rightarrow \mathbb{R}$ définie par : 

    $$
    \forall x\in \mathbb{R}, \text{ }f_0(x) = sin(x)
    $$
    
    Nous considérons le triplet $(x_1,\text{ } x_2,\text{ } x_3) = (-\frac{\pi}{2},\text{ } -\frac{\pi}{2} + \frac{1}{2},\text{ } \frac{\pi}{2})$ dont chaque élément va servir de point initial pour l'algorithme de Newton.

    La fonction $f_0$ est bien dérivable (et même infiniment dérivable) est sa dérivée est définie par : $\forall x\in \mathbb{R}, \text{ }f'_0(x) = cos(x)$ 

    L'algorithme de Newton cherche alors les points où la dérivée s'annule (et plus généralement, pour les fonctions vectorielles, les points où le gradient s'annule), or on remarque rapidement que $f'_0(x_1) = f'_0(x_3) = 0$ et par conséquent l'algorithme se termine sans faire une itération étant donné que le point initial est directement la solution (c'est-à-dire $x_{1_{sol}} = x_1$ et $x_{3_{sol}} = x_3$). 
    
    Le drapeau levé par l'algorithme est alors celui de la convergence de la solution (c'est-à-dire 0) et bien évidemment, on a : $f_0(x_{1_{sol}}) = -1$

    Mais il faut faire attention, l'algorithme de Newton cherche les points critiques d'une fonction et non forcément le minimum, dans notre cas $x_3$ est un maximum de la fonction $f_0$ et on a bien $f_0(x_{3_{sol}}) = 1$ 

    Quant à $x_2$, il faut faire quelques itérations avant d'atteindre la solution (et plus exactement 3 itérations), en effet comme $f_0'(x_2)\neq 0$, une première itération est donnée par : 

    $$
    x_{2_1} = x_2 - \frac{f_0'(x_2)}{f_0''(x_2)} = x_2 + \frac{cos(x_2)}{sin(x_2)} = x_2 + cotan(x_2) \text{  avec $f_0''(x) = -sin(x)$ qui est bien définie}
    $$

    Finalement, après 3 itérations, l'algorithme de Newton trouve $x_{2_{sol}} = -\frac{\pi}{2} = x_1$ qui est bien un point critique de la fonction $f_0$ et lève bien un drapeau indiquant la convergence de la solution.

2. Soit la fonction $f_1 : \mathbb{R}^3 \rightarrow \mathbb{R}$ définie par : 

    $$
    \forall (x_1, x_2, x_3)\in \mathbb{R}^3, \text{ } f_1(x_1, x_2, x_3) = 2(x_1 + x_2 + x_3 - 3)^2 + (x_1 - x_2)^2 + (x_2 - x_3)^2
    $$ 

   La fonction $f_1$ est bien définie et différentiable au moins deux fois, déterminons son gradient $\nabla f_1$ et sa hessienne $\nabla^2 f_1$ : 

    $$
    \nabla f_1(x_1, x_2, x_3) = \begin{pmatrix}
                                6x_1+2x_2+4x_3-12\\
                                2x_1+8x_2+2x_3-12\\
                                4x_1+2x_2+6x_1-12\\
                                \end{pmatrix} 
    \hskip{0.5cm}\text{et}\hskip{0.5cm}
    \nabla ^ 2 f_1(x_1, x_2, x_3) = \begin{pmatrix}
                                    6 & 2 & 4\\
                                    2 & 8 & 2\\
                                    4 & 2 & 6\\
                                    \end{pmatrix} 
    $$

    Or $\forall (x_1, x_2, x_3)\in \mathbb{R}^3, \text{ } \nabla^2 f_1(x_1, x_2, x_3)$ est symétrique **définie positive**, la méthode de Newton revient alors à minimiser le modèle quadratique de la fonction $f_1$ noté $m_1$.

    Mais étant donné que $m_1$ est quadratique, son gradient $\nabla m_1$ a alors des coordonnées affine, il suffit alors d'une itération pour obtenir le 0 de $\nabla m_1$ (cela revient en une dimension à chercher le zéro d'une fonction affine, c'est son intersection avec la droite des abscisses). L'algorithme de Newton se termine alors forcément en zéro itération si le point initial est pas la solution et en une itération sinon

3. Soit la fonction $f_2 : \mathbb{R}^2 \rightarrow \mathbb{R}$ définie par : 

    $$
    \forall (x_1, x_2)\in \mathbb{R}^2, \text{ } f_2(x_1, x_2) = 100(x_2 - x_1)^2 + (1-x_1)^2
    $$

    Il s'agit de la fonction de Rosenbrock, elle est différentiable au moins deux fois, déterminons son gradient $\nabla f_2$ et sa hessienne $\nabla^2 f_2$

    $$
    \nabla f_1(x_1, x_2) = \begin{pmatrix}
                            400x_1^3 - 400x_2x_1 + 2x_1 - 2\\
                            -200x_1^2 + 200x_2\\
                            \end{pmatrix} 
    \hskip{0.5cm}\text{et}\hskip{0.5cm}
    \nabla ^ 2 f_1(x_1, x_2) = \begin{pmatrix}
                                1200x_1^2 - 400x_2 + 2 & -400x_1 \\
                                -400x_1 & 200 \\
                                \end{pmatrix} 
    $$

    Or, si à l'itération $k$ de l'algorithme de Newton nous obtenons l'itéré $x_k$, alors à l'itéré $x_{k+1}$ est obtenu à l'itération $k+1$ à partir de la somme entre $x_k$ et de $d_k$. 
    
    $d_k$ est obtenu en résolvant le système linéaire suivant : 

    $$
    \nabla^2 f_2(x_k) \cdot d_k = - \nabla f_2(x_k) \hskip{1cm} (1)
    $$ 
    
    Soit $\epsilon > 0$, on considère $\bar{x}=(0,\epsilon)$, nous obtenons alors : 

    $$
    \nabla f_1(\bar{x}) = \begin{pmatrix}
                            2\\
                            200\epsilon\\
                            \end{pmatrix} 
    \hskip{0.5cm}\text{et}\hskip{0.5cm}
    \nabla ^ 2 f_1(\bar{x}) = \begin{pmatrix}
                                    2 - 400\epsilon & 0 \\
                                    0 & 200 \\
                                    \end{pmatrix} 
    $$

    La résolution de $(1)$ avec $\bar{x}$ donne alors $\bar{d} =  \begin{pmatrix}
                                                                  \frac{2}{2-400\epsilon}\\
                                                                  \epsilon\\
                                                                  \end{pmatrix}$
    et donc pour $\epsilon \rightarrow \frac{2}{400} = 0.005$, $\hskip{0.1cm}$ $\bar{d_1} \rightarrow +\infty$.

    Par conséquent, $f_2$ ne peut pas converger pour cette famille de points initiaux.
